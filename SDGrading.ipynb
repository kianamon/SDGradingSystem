```python
#Libraries in use:
import pandas as pd
import csv
import re
import numpy as np
```

```python
gc_df = pd.read_csv("./gc.csv")
del gc_df['Weighted Total [Total Pts: up to 0 Percentage] |1649396']
del gc_df['Last Access']
del gc_df['Availability']
gc_df['Full Name'] = gc_df['First Name'] +" "+ gc_df['Last Name']
gc_df["Department"] = "ECE"
gc = gc_df.drop(['First Name', 'Last Name', 
                 'Final Written Progress report [Total Pts: 30 Score] |1738697',
                 'Final Oral Presentation [Total Pts: 30 Score] |1738698'], axis=1)
gc.head()
```

```python
Team_df = pd.read_csv("./TeamResults.csv")
Team_df['Last Name'] = Team_df['Q2'].str.split(',', expand=True)[0]
Team_df['First Name'] = Team_df['Q2'].str.split(',', expand=True)[1]
del Team_df['Q2']
Team_df['Full Name'] = Team_df['First Name'] +" "+ Team_df['Last Name']
Team_df['Full Name'] = Team_df['Full Name'].astype(str).map(lambda x: x.strip())
Team = Team_df.drop(['First Name', 'Last Name'], axis=1)
Team.head()
```

```python
new1 = pd.merge(
        gc, Team,  how='outer', on='Full Name')
new1 = new1.rename(columns={'AverageGrade': 'TeamWork Grade'})
#new11 = new1.drop(['Unnamed: 0'], axis = 1)
new1.info()
```

```python
groups_df = pd.read_csv("./groups.csv")
groups_df['Group ID'] = groups_df['Group Code'].apply(lambda x: re.search(r'\d+', x).group())
groups_df["TeamID"] = pd.to_numeric(groups_df['Group ID'], downcast="float")
group = groups_df.drop(['Group Code', 'Group ID'], axis=1)
group['Full Name'] = group['First Name'] +" "+ group['Last Name']
group['Full Name'] = group['Full Name'].astype(str).map(lambda x: x.strip())
grp = group.drop(['First Name', 'Last Name'], axis=1)
grp.head()
```

```python
new2 = pd.merge(
        new1, grp,  how='outer', on='Full Name')
new2 = new2.rename(columns={'TeamID': 'GroupID'})
new3 = new2.drop(['Student Id', 'User Name'], axis=1)
new3.head()
```

```python
new3.to_csv("new3.csv", sep=',')
```

```python
new33 = pd.read_csv("./new3.csv")
```

```python
oral_df = pd.read_csv("./OralResults.csv")
oral_df['GroupID'] = pd.to_numeric(oral_df['TeamID'], downcast = "float")
oral = oral_df.drop(['TeamID'], axis=1)
oral.head()
```

```python
new4 = pd.merge(new33, oral,  how='outer', on='GroupID')
new4 = new4.rename(columns={'Q3': 'Project Title', 'AverageGrade': 'Oral Presentation Grade', 
                            'Graders_y': 'Oral Pres. Graders',
                            'Graders_x': 'Teamwork Grader'})
new44 = new4.drop(['Unnamed: 0_x', 'Unnamed: 0_y'], axis = 1)
new44.head()
```

```python
new44.to_csv("new44.csv", sep=',')
```

```python
written_df = pd.read_csv("./WrittenResults.csv")
written_df['GroupID'] = pd.to_numeric(written_df['TeamID'], downcast = "float")
written = written_df.drop(['TeamID'], axis=1)
written.head()
```

```python
new5 = pd.merge(new4, written,  how='outer', on='GroupID')
new5 = new5.rename(columns={'AverageGrade': 'Written Presentation Grade', 'Graders': 'Written Report Graders'})
new55 = new5.drop(['Unnamed: 0_x', 'Unnamed: 0_y'], axis = 1)
new55.head()
```

```python
new55.to_csv("new55.csv", sep=',')
```
